```python
# Parity Bit Implementation

# Cell 1: Import required libraries and define functions
import random

def add_even_parity(data):
    """
    Add even parity bit to binary string
    Even parity: parity bit makes total number of 1s even
    """
    ones_count = data.count('1')
    parity_bit = '0' if ones_count % 2 == 0 else '1'
    return data + parity_bit

def add_odd_parity(data):
    """
    Add odd parity bit to binary string
    Odd parity: parity bit makes total number of 1s odd
    """
    ones_count = data.count('1')
    parity_bit = '1' if ones_count % 2 == 0 else '0'
    return data + parity_bit

def verify_even_parity(data_with_parity):
    """
    Verify even parity bit
    Returns True if parity is correct, False if error detected
    """
    ones_count = data_with_parity.count('1')
    return ones_count % 2 == 0

def verify_odd_parity(data_with_parity):
    """
    Verify odd parity bit
    Returns True if parity is correct, False if error detected
    """
    ones_count = data_with_parity.count('1')
    return ones_count % 2 == 1

def introduce_error(data, num_errors=1):
    """
    Introduce random bit errors for testing
    """
    data_list = list(data)
    positions = random.sample(range(len(data)), min(num_errors, len(data)))

    for pos in positions:
        data_list[pos] = '0' if data_list[pos] == '1' else '1'

    return ''.join(data_list), positions

print("✓ Parity bit functions defined successfully!")

# Cell 2: Test with sample data - Even Parity
print("=== EVEN PARITY TESTING ===")
print()

# Test data
test_data = [
    "1011",      # 3 ones
    "1100",      # 2 ones
    "1111",      # 4 ones
    "0000",      # 0 ones
    "10101010"   # 4 ones
]

for data in test_data:
    # Sender side - add parity
    data_with_parity = add_even_parity(data)
    ones_in_original = data.count('1')

    print(f"Original data: {data} (ones: {ones_in_original})")
    print(f"With parity:   {data_with_parity} (parity bit: {data_with_parity[-1]})")

    # Receiver side - verify parity
    is_valid = verify_even_parity(data_with_parity)
    print(f"Parity check:  {'✓ VALID' if is_valid else '✗ ERROR DETECTED'}")
    print("-" * 50)

# Cell 3: Test with sample data - Odd Parity
print("=== ODD PARITY TESTING ===")
print()

for data in test_data:
    # Sender side - add parity
    data_with_parity = add_odd_parity(data)
    ones_in_original = data.count('1')

    print(f"Original data: {data} (ones: {ones_in_original})")
    print(f"With parity:   {data_with_parity} (parity bit: {data_with_parity[-1]})")

    # Receiver side - verify parity
    is_valid = verify_odd_parity(data_with_parity)
    print(f"Parity check:  {'✓ VALID' if is_valid else '✗ ERROR DETECTED'}")
    print("-" * 50)

# Cell 4: Test with intentional errors - Even Parity
print("=== ERROR DETECTION TESTING (Even Parity) ===")
print()

test_data_for_errors = ["1011", "1100", "10101010"]

for data in test_data_for_errors:
    print(f"Testing with data: {data}")

    # Add parity bit
    data_with_parity = add_even_parity(data)
    print(f"Original with parity: {data_with_parity}")

    # Test 1: No error
    is_valid = verify_even_parity(data_with_parity)
    print(f"No error - Parity check: {'✓ VALID' if is_valid else '✗ ERROR'}")

    # Test 2: Single bit error
    corrupted_data, error_positions = introduce_error(data_with_parity, 1)
    is_valid = verify_even_parity(corrupted_data)
    print(f"1 bit error at position {error_positions[0]}: {corrupted_data}")
    print(f"Parity check: {'✓ VALID' if is_valid else '✗ ERROR DETECTED'}")

    # Test 3: Two bit errors (parity cannot detect even number of errors)
    corrupted_data, error_positions = introduce_error(data_with_parity, 2)
    is_valid = verify_even_parity(corrupted_data)
    print(f"2 bit errors at positions {error_positions}: {corrupted_data}")
    print(f"Parity check: {'✓ VALID' if is_valid else '✗ ERROR DETECTED'}")
    print("Note: Even number of errors may not be detected by parity bit!")
    print("-" * 60)

# Cell 5: Interactive testing function
def interactive_parity_test():
    """
    Interactive function to test parity with custom data
    """
    print("=== INTERACTIVE PARITY TESTING ===")

    while True:
        data = input("\nEnter binary string (or 'quit' to exit): ").strip()

        if data.lower() == 'quit':
            break

        if not all(bit in '01' for bit in data):
            print("Please enter only binary digits (0 and 1)")
            continue

        if len(data) == 0:
            print("Please enter at least one bit")
            continue

        print(f"\nOriginal data: {data}")

        # Even parity
        even_parity_data = add_even_parity(data)
        print(f"Even parity:   {even_parity_data}")
        print(f"Verification:  {'✓ VALID' if verify_even_parity(even_parity_data) else '✗ ERROR'}")

        # Odd parity
        odd_parity_data = add_odd_parity(data)
        print(f"Odd parity:    {odd_parity_data}")
        print(f"Verification:  {'✓ VALID' if verify_odd_parity(odd_parity_data) else '✗ ERROR'}")

        # Test with error
        print("\nTesting with 1-bit error:")
        corrupted_even, pos = introduce_error(even_parity_data, 1)
        print(f"Corrupted even parity: {corrupted_even} (error at position {pos[0]})")
        print(f"Error detection: {'✓ DETECTED' if not verify_even_parity(corrupted_even) else '✗ NOT DETECTED'}")

print("✓ Interactive testing function ready!")
print("Run: interactive_parity_test() to start interactive testing")

# Cell 6: Summary and explanation
print("=== PARITY BIT EXPLANATION ===")
print()
print("EVEN PARITY:")
print("- Parity bit is set so total number of 1s is even")
print("- If data has even number of 1s → parity bit = 0")
print("- If data has odd number of 1s → parity bit = 1")
print()
print("ODD PARITY:")
print("- Parity bit is set so total number of 1s is odd")
print("- If data has even number of 1s → parity bit = 1")
print("- If data has odd number of 1s → parity bit = 0")
print()
print("ERROR DETECTION:")
print("✓ Can detect single bit errors")
print("✓ Can detect any odd number of bit errors")
print("✗ Cannot detect even number of bit errors")
print("✗ Cannot correct errors, only detect them")
print()
print("LIMITATION:")
print("If 2, 4, 6, ... bits are corrupted, parity check may still pass!")
```
