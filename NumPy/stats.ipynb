```python
import numpy as np
from scipy.stats import ttest_1samp, shapiro
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (product weights)
data = np.array([63, 75, 84, 58, 52, 96, 63, 55, 76, 83])

# Check normality
stat, p = shapiro(data)
print("Shapiro-Wilk p-value:", p)  # If p < 0.05, data is non-normal

# One-sample t-test
t_stat, p_value = ttest_1samp(data, 68)
print(f"t-statistic: {t_stat:.2f}, p-value: {p_value:.4f}")
print("Hypothesis", "Rejected" if p_value < 0.05 else "Accepted")

# Visualize
sns.histplot(data, kde=True)
plt.axvline(68, color='red', linestyle='--', label='Hypothesized Mean (68)')
plt.axvline(np.mean(data), color='green', label='Sample Mean')
plt.legend()
plt.title("Product Weights Distribution")
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_ind, levene
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (two groups of student weights)
data1 = np.array([63, 75, 84, 58, 52, 96, 63, 55, 76, 83])
data2 = np.array([53, 43, 31, 113, 33, 57, 27, 23, 24, 43])

# Check equal variances
stat, p_levene = levene(data1, data2)
print("Levene's test p-value:", p_levene)  # If p < 0.05, variances unequal

# Two-sample t-test
stat, p = ttest_ind(data1, data2, equal_var=p_levene > 0.05)
print(f"t-statistic: {stat:.2f}, p-value: {p:.4f}")
print("Hypothesis", "Rejected" if p < 0.05 else "Accepted")

# Visualize
sns.boxplot(data=[data1, data2])
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.title("Weight Comparison Between Groups")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, f_oneway, shapiro, levene

# Simulate HR dataset
np.random.seed(42)
data = pd.DataFrame({
    'employee_id': range(1, 101),
    'performance_score': np.random.normal(80, 10, 100),
    'satisfaction_score': np.random.normal(7, 1.5, 100),
    'tenure_years': np.random.exponential(3, 100),
    'location': np.random.choice(['Mumbai', 'Chicago', 'London'], 100)
})

# 1. Descriptive Statistics
print("Skewness:\\n", data[['performance_score', 'satisfaction_score', 'tenure_years']].skew())
print("\\nKurtosis:\\n", data[['performance_score', 'satisfaction_score', 'tenure_years']].kurtosis())

# Visualize distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(['performance_score', 'satisfaction_score', 'tenure_years']):
    sns.histplot(data[col], kde=True, ax=axes[i])
    axes[i].set_title(f"{col} (Skew: {data[col].skew():.2f}, Kurt: {data[col].kurtosis():.2f})")
plt.tight_layout()
plt.show()

# 2. Correlation Analysis
corr_matrix = data[['performance_score', 'satisfaction_score', 'tenure_years']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()

# 3. Hypothesis Testing: Compare performance by location
locations = data['location'].unique()
groups = [data[data['location'] == loc]['performance_score'] for loc in locations]

# Check assumptions
stat, p_levene = levene(*groups)
print("Levene's test p-value:", p_levene)
for group, loc in zip(groups, locations):
    stat, p_shapiro = shapiro(group)
    print(f"Shapiro-Wilk p-value for {loc}:", p_shapiro)

# ANOVA
stat, p = f_oneway(*groups)
print(f"ANOVA F-statistic: {stat:.2f}, p-value: {p:.4f}")
print("Hypothesis", "Rejected" if p < 0.05 else "Accepted")

# Visualize
sns.boxplot(x='location', y='performance_score', data=data)
plt.title("Performance Scores by Location")
plt.show()
```

```python
# Let's build a more realistic example
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Simulating real customer data
np.random.seed(42)
n_customers = 1000

# Create realistic customer segments and purchase behavior
customer_data = pd.DataFrame({
    'age_group': np.random.choice(['18-25', '26-35', '36-45', '46+'], n_customers,
                                 p=[0.2, 0.3, 0.3, 0.2]),
    'purchased_premium': np.random.choice(['Yes', 'No'], n_customers, p=[0.3, 0.7])
})

# Create contingency table
contingency = pd.crosstab(customer_data['age_group'],
                         customer_data['purchased_premium'])
print("Contingency Table:")
print(contingency)

# Perform Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency)

print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")
```

```python
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt

# Real scenario: Comparing customer satisfaction between two service channels
# Email support vs. Phone support (satisfaction scores 1-10)
email_satisfaction = [6, 7, 5, 8, 7, 6, 9, 5, 7, 8, 6, 7]
phone_satisfaction = [8, 9, 7, 9, 8, 9, 10, 8, 9, 7, 8, 9]

# Visualize first
plt.figure(figsize=(10, 6))
plt.boxplot([email_satisfaction, phone_satisfaction],
            labels=['Email Support', 'Phone Support'])
plt.ylabel('Satisfaction Score')
plt.title('Customer Satisfaction by Support Channel')
plt.show()

# Mann-Whitney U test
statistic, p_value = mannwhitneyu(email_satisfaction, phone_satisfaction,
                                  alternative='two-sided')

print(f"Mann-Whitney U statistic: {statistic}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference between support channels")
else:
    print("No significant difference between support channels")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, mannwhitneyu, wilcoxon, kruskal
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate customer satisfaction dataset
np.random.seed(42)
data = pd.DataFrame({
    'customer_id': range(1, 201),
    'satisfaction_score': np.random.randint(1, 6, 200),  # Ordinal (1-5)
    'region': np.random.choice(['North', 'South', 'East'], 200),
    'pre_campaign_score': np.random.randint(1, 6, 200),
    'post_campaign_score': np.random.randint(1, 6, 200)
})

# 1. Chi-Square Test: Satisfaction vs. Region
contingency_table = pd.crosstab(data['region'], data['satisfaction_score'])
stat, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-Square Test: Satisfaction vs. Region")
print(f"Statistic: {stat:.2f}, p-value: {p:.4f}, Degrees of Freedom: {dof}")
print("Hypothesis", "Rejected" if p < 0.05 else "Accepted")
sns.heatmap(contingency_table, annot=True, cmap='Blues', fmt='d')
plt.title("Contingency Table: Satisfaction vs. Region")
plt.show()

# 2. Mann-Whitney U Test: Satisfaction scores between two regions
north_scores = data[data['region'] == 'North']['satisfaction_score']
south_scores = data[data['region'] == 'South']['satisfaction_score']
stat, p = mannwhitneyu(north_scores, south_scores, alternative='two-sided')
print("\\nMann-Whitney U Test: North vs. South Satisfaction")
print(f"Statistic: {stat:.2f}, p-value: {p:.4f}")
print("Hypothesis", "Rejected" if p < 0.01 else "Accepted")
sns.boxplot(x='region', y='satisfaction_score', data=data[data['region'].isin(['North', 'South'])])
plt.title("Satisfaction Scores: North vs. South")
plt.show()

# 3. Wilcoxon Test: Pre- vs. Post-Campaign Scores
stat, p = wilcoxon(data['pre_campaign_score'], data['post_campaign_score'])
print("\\nWilcoxon Test: Pre- vs. Post-Campaign Scores")
print(f"Statistic: {stat:.2f}, p-value: {p:.4f}")
print("Hypothesis", "Rejected" if p < 0.01 else "Accepted")
sns.boxplot(data=[data['pre_campaign_score'], data['post_campaign_score']])
plt.xticks([0, 1], ['Pre-Campaign', 'Post-Campaign'])
plt.title("Pre- vs. Post-Campaign Satisfaction Scores")
plt.show()

# 4. Kruskal-Wallis Test: Satisfaction across all regions
groups = [data[data['region'] == r]['satisfaction_score'] for r in data['region'].unique()]
stat, p = kruskal(*groups)
print("\\nKruskal-Wallis Test: Satisfaction across Regions")
print(f"Statistic: {stat:.2f}, p-value: {p:.4f}")
print("Hypothesis", "Rejected" if p < 0.01 else "Accepted")
sns.boxplot(x='region', y='satisfaction_score', data=data)
plt.title("Satisfaction Scores by Region")
plt.show()
```
