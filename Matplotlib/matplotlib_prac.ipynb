```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, 'r--', label='sin(x)')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))
def update(frame):
    line.set_ydata(np.sin(x + frame/10))
    return line,
ani = animation.FuncAnimation(fig, update, frames=100, interval=50)
plt.show()
```

```python
from matplotlib.axes import Axes
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

class CustomAxes(Axes):
    def plot(self, *args, **kwargs):
        super().plot(*args, **kwargs)
        self.set_title('Custom Plot')

fig = Figure(figsize=(6, 4)) # Create a figure
ax = CustomAxes(fig, [0.1, 0.1, 0.8, 0.8]) # Create an instance of your custom axes
fig.add_axes(ax) # Add the custom axes to the figure

ax.plot([1, 2, 3], [4, 5, 6])
plt.show(fig) # Use plt.show(fig) to display the specific figure
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
# Download stock data
data = yf.download('AAPL', start='2024-01-01', end='2025-01-01')
# Calculate moving averages
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()
# Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data['Close'], label='Close Price', color='blue')
ax.plot(data['MA50'], label='50-Day MA', color='orange')
ax.plot(data['MA200'], label='200-Day MA', color='red')
ax.set_title('AAPL Stock Price with Moving Averages')
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.legend()
ax.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
# Initialize temperature grid
grid = np.random.rand(50, 50)
fig, ax = plt.subplots()
im = ax.imshow(grid, cmap='hot')
def update(frame):
    global grid
    grid = grid + np.random.randn(50, 50) * 0.1
    im.set_array(grid)
    return [im]
ani = animation.FuncAnimation(fig, update, frames=100, interval=100, blit=True)
plt.colorbar(im)
plt.show()
```
